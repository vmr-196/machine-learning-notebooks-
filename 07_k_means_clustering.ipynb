# K-Means Clustering

A beginner-friendly implementation of K-Means Clustering using Scikit-Learn.

**Objective:** Group customers into different clusters based on Age and Annual Income.

## 1. Introduction

K-Means Clustering is an Unsupervised Machine Learning algorithm.

Unlike classification algorithms, K-Means does not require labels.

The algorithm automatically groups similar data points into clusters.

### Applications

- Customer Segmentation
- Market Analysis
- Recommendation Systems
- Image Compression
- Fraud Detection

## 2. How K-Means Works

K-Means follows these steps:

1. Select K cluster centers (Centroids)
2. Assign each data point to the nearest centroid
3. Recalculate centroid positions
4. Repeat until centroids stop moving

### Example

Cluster 1 → Students

Cluster 2 → Working Professionals

Cluster 3 → Business Owners

```python
import numpy as np
import matplotlib.pyplot as plt
import joblib

from sklearn.cluster import KMeans
```

## 3. Creating the Dataset

Dataset Features:

- Age
- Annual Income

No target column is required because K-Means is an unsupervised algorithm.

```python
X = np.array([
    [20, 25000],
    [22, 28000],
    [25, 30000],
    [30, 35000],
    [35, 60000],
    [38, 62000],
    [40, 65000],
    [42, 68000],
    [50, 90000],
    [55, 95000]
])

print("Dataset Loaded Successfully")
```

## 4. Visualizing the Dataset

Each point represents one customer.

```python
plt.scatter(X[:,0], X[:,1])

plt.xlabel("Age")
plt.ylabel("Annual Income")

plt.title("Customer Dataset")

plt.show()
```

## 5. Finding the Optimal Number of Clusters

The Elbow Method helps determine the best value of K.

```python
wcss = []

for i in range(1, 6):
    kmeans = KMeans(
        n_clusters=i,
        random_state=42,
        n_init=10
    )

    kmeans.fit(X)

    wcss.append(kmeans.inertia_)

plt.plot(range(1, 6), wcss, marker='o')

plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")

plt.title("Elbow Method")

plt.show()
```

## 6. Creating the K-Means Model

We choose K = 3 based on the Elbow Method.

```python
model = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)
```

## 7. Training the Model

The fit() method identifies cluster centers.

```python
model.fit(X)

print("Model Trained Successfully")
```

## 8. Predicting Clusters

The model assigns a cluster number to each customer.

```python
clusters = model.predict(X)

print(clusters)
```

## 9. Visualizing the Clusters

Each color represents a different cluster.

The red X marks the cluster centroids.

```python
clusters = model.labels_

plt.scatter(
    X[:,0],
    X[:,1],
    c=clusters
)

plt.scatter(
    model.cluster_centers_[:,0],
    model.cluster_centers_[:,1],
    marker='X',
    s=200
)

plt.xlabel("Age")
plt.ylabel("Annual Income")

plt.title("Customer Clusters")

plt.show()
```

## 10. Saving the Model

The trained K-Means model can be saved and reused later.

```python
joblib.dump(model, "kmeans_model.pkl")

print("Model Saved Successfully")
```

## 11. Loading the Saved Model

The saved model can be loaded without retraining.

```python
loaded_model = joblib.load("kmeans_model.pkl")

result = loaded_model.predict([[28, 32000]])

print("Assigned Cluster:", result[0])
```
