# Random Forest Classifier

A beginner-friendly implementation of Random Forest using Scikit-Learn.

**Objective:** Predict whether a customer will buy a car based on age and income.

## 1. Introduction

Random Forest is a supervised machine learning algorithm used for classification and regression.

It combines multiple Decision Trees and makes predictions based on majority voting.

### Applications

- Loan Approval
- Fraud Detection
- Disease Prediction
- Customer Churn Prediction
- Stock Market Analysis

## 2. How Random Forest Works

Random Forest creates multiple Decision Trees.

Each tree makes its own prediction.

The final prediction is decided by majority voting.

Example:

Tree 1 → Buy

Tree 2 → Buy

Tree 3 → Don't Buy

Final Prediction → Buy

```python
import numpy as np
import joblib

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

## 3. Creating the Dataset

Dataset Features:

- Age
- Income

Target:

- Buy Car

0 = No
1 = Yes

```python
X = np.array([
    [22, 25000],
    [25, 30000],
    [28, 35000],
    [35, 60000],
    [40, 65000],
    [45, 70000],
    [23, 27000],
    [38, 62000]
])

y = np.array([
    0,
    0,
    0,
    1,
    1,
    1,
    0,
    1
])

print("Dataset Loaded Successfully")
```

## 4. Creating the Random Forest Model

We will use 100 Decision Trees.

n_estimators = Number of Trees

```python
model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)
```

## 5. Training the Model

The fit() method trains all Decision Trees in the forest.

```python
model.fit(X, y)

print("Model Trained Successfully")
```

## 6. Making Predictions

Let's predict whether a customer aged 30 with an income of ₹55,000 will buy a car.

```python
prediction = model.predict([[30, 55000]])

print("Prediction:", prediction[0])

if prediction[0] == 1:
    print("Customer Will Buy Car")
else:
    print("Customer Will Not Buy Car")
```

## 7. Evaluating Model Performance

Accuracy measures how many predictions are correct.

```python
train_predictions = model.predict(X)

accuracy = accuracy_score(y, train_predictions)

print("Accuracy:", accuracy * 100, "%")
```

## 8. Feature Importance

Random Forest can show which features are most important.

```python
for feature, importance in zip(
    ["Age", "Income"],
    model.feature_importances_
):
    print(feature, ":", round(importance, 3))
```

## 9. Saving the Model

The trained model can be saved and reused later.

```python
joblib.dump(model, "random_forest.pkl")

print("Model Saved Successfully")
```

## 10. Loading the Saved Model

The saved model can be loaded without retraining.

```python
loaded_model = joblib.load("random_forest.pkl")

result = loaded_model.predict([[32, 58000]])

print(result)
```
