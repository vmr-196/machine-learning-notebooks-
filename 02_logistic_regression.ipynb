# Logistic Regression

A beginner-friendly implementation of Logistic Regression using Scikit-Learn.

**Objective:** Predict whether a student will Pass or Fail based on study hours.

## 1. Introduction

Logistic Regression is a supervised machine learning algorithm used for classification problems.

Unlike Linear Regression, which predicts numerical values, Logistic Regression predicts categories.

### Applications

- Spam Email Detection
- Disease Prediction
- Loan Approval
- Pass/Fail Prediction
- Customer Churn Prediction

## 2. How Logistic Regression Works

Logistic Regression calculates the probability of an event occurring.

If probability > 0.5:
    Class = 1

Else:
    Class = 0

In this notebook:

0 = Fail
1 = Pass

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import joblib
```

## 3. Creating the Dataset

We will use a simple dataset containing:

- Study Hours
- Pass or Fail

The model will learn how study hours affect exam results.

```python
X = np.array([
    [1],
    [2],
    [3],
    [4],
    [5],
    [6],
    [7],
    [8]
])

y = np.array([
    0,
    0,
    0,
    0,
    1,
    1,
    1,
    1
])

print("Dataset Loaded Successfully")
```

## 4. Visualizing the Dataset

Each point represents a student.

0 = Fail

1 = Pass

```python
plt.scatter(X, y)

plt.xlabel("Study Hours")
plt.ylabel("Result")

plt.title("Study Hours vs Result")

plt.show()
```

## 5. Creating the Logistic Regression Model

Now we create a Logistic Regression model that will classify students as Pass or Fail.

```python
model = LogisticRegression()
```

## 6. Training the Model

The fit() method trains the model using the dataset.

```python
model.fit(X, y)

print("Model Trained Successfully")
```

## 7. Making Predictions

Let's predict whether a student studying for 5.5 hours will pass.

```python
prediction = model.predict([[5.5]])

print("Prediction:", prediction[0])

if prediction[0] == 1:
    print("Student Will Pass")
else:
    print("Student Will Fail")
```

## 8. Checking Probability

Logistic Regression predicts probabilities before assigning classes.

```python
probability = model.predict_proba([[5.5]])

print(probability)
```

## 9. Evaluating Model Performance

Accuracy measures how many predictions are correct.

```python
train_predictions = model.predict(X)

accuracy = accuracy_score(y, train_predictions)

print("Accuracy:", accuracy * 100, "%")
```

## 10. Saving the Model

The trained model can be saved for future use.

```python
joblib.dump(model, "logistic_regression.pkl")

print("Model Saved Successfully")
```

## 11. Loading the Saved Model

We can load the saved model and make predictions without retraining.

```python
loaded_model = joblib.load("logistic_regression.pkl")

result = loaded_model.predict([[6]])

print(result)
```
