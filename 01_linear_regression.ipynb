# Linear Regression

A beginner-friendly implementation of Linear Regression using Scikit-Learn.

**Objective:** Predict house prices based on house size.

## 1. Introduction

Linear Regression is a supervised machine learning algorithm used to predict continuous numerical values.

### Applications
- House Price Prediction
- Salary Prediction
- Sales Forecasting
- Stock Trend Analysis

### Why Use Linear Regression?

It helps us understand the relationship between input variables and output variables by fitting the best straight line through the data.

## 2. Mathematical Formula

Linear Regression follows the equation:

y = mx + b

Where:

- y = Predicted value
- m = Slope of line
- x = Input feature
- b = Intercept

The algorithm finds the best values of m and b that minimize prediction errors.

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
import joblib
```

## 3. Creating the Dataset

We will create a simple dataset containing:

- House Size (sq ft)
- House Price (Lakhs)

The model will learn how house size affects price.

```python
X = np.array([
    [500],
    [700],
    [1000],
    [1200],
    [1500],
    [1700],
    [2000]
])

y = np.array([
    10,
    15,
    20,
    24,
    30,
    34,
    40
])

print("Dataset Loaded Successfully")
```

## 4. Visualizing the Dataset

A scatter plot helps us understand the relationship between house size and price.

Each dot represents one house.

```python
plt.scatter(X, y)

plt.xlabel("House Size (sq ft)")
plt.ylabel("Price (Lakhs)")
plt.title("House Size vs Price")

plt.show()
```

## 5. Creating the Linear Regression Model

Now we create a Linear Regression object that will learn patterns from the dataset.

```python
model = LinearRegression()
```

## 6. Training the Model

The fit() function trains the model using the available dataset.

The model learns the relationship between house size and price.

```python
model.fit(X, y)

print("Model Trained Successfully")
```

## 7. Making Predictions

Let's predict the price of a house with an area of 1800 sq ft.

```python
pred = model.predict([[1800]])

print("Predicted Price:", round(pred[0], 2), "Lakhs")
```

## 8. Evaluating Model Performance

Mean Absolute Error (MAE) measures the average prediction error.

Lower MAE indicates better performance.

```python
train_pred = model.predict(X)

mae = mean_absolute_error(y, train_pred)

print("Mean Absolute Error:", round(mae, 3))
```

## 9. Visualizing the Regression Line

The blue dots represent actual data.

The line represents predictions made by the model.

```python
plt.scatter(X, y)

plt.plot(X, train_pred)

plt.xlabel("House Size (sq ft)")
plt.ylabel("Price (Lakhs)")
plt.title("Linear Regression Best Fit Line")

plt.show()
```

## 10. Saving the Trained Model

Machine learning models can be saved and reused later without retraining.

We use Joblib to save the model.

```python
joblib.dump(model, "linear_regression.pkl")

print("Model Saved Successfully")
```

## 11. Loading the Saved Model

The saved model can be loaded and used for future predictions.

```python
loaded_model = joblib.load("linear_regression.pkl")

prediction = loaded_model.predict([[2200]])

print("Predicted Price:", prediction[0])
```
