# Support Vector Machine (SVM)

A beginner-friendly implementation of Support Vector Machine using Scikit-Learn.

**Objective:** Predict whether a customer will buy a product based on age and income.

## 1. Introduction

Support Vector Machine (SVM) is a supervised machine learning algorithm used for classification and regression tasks.

SVM works by finding the best possible boundary (hyperplane) that separates different classes.

### Applications

- Face Recognition
- Image Classification
- Spam Detection
- Handwriting Recognition
- Medical Diagnosis

## 2. How SVM Works

SVM tries to find the optimal boundary between classes.

The boundary should maximize the distance between classes.

This distance is called the Margin.

### Example

Class A ● ● ●

------------ Best Boundary ------------

Class B ▲ ▲ ▲

The larger the margin, the better the classification.

```python
import numpy as np
import matplotlib.pyplot as plt
import joblib

from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

## 3. Creating the Dataset

Dataset Features:

- Age
- Income

Target:

- Buy Product

0 = No
1 = Yes

```python
X = np.array([
    [22, 25000],
    [25, 30000],
    [28, 35000],
    [35, 60000],
    [40, 65000],
    [45, 70000],
    [23, 27000],
    [38, 62000]
])

y = np.array([
    0,
    0,
    0,
    1,
    1,
    1,
    0,
    1
])

print("Dataset Loaded Successfully")
```

## 4. Visualizing the Dataset

Let's visualize customer data using a scatter plot.

```python
plt.scatter(X[:,0], X[:,1])

plt.xlabel("Age")
plt.ylabel("Income")
plt.title("Customer Dataset")

plt.show()
```

## 5. Creating the SVM Model

We use SVC (Support Vector Classifier) for classification tasks.

```python
model = SVC(kernel='linear')
```

## 6. Training the Model

The fit() method trains the SVM model using the dataset.

```python
model.fit(X, y)

print("Model Trained Successfully")
```

## 7. Making Predictions

Let's predict whether a customer aged 30 with an income of ₹55,000 will buy the product.

```python
prediction = model.predict([[30, 55000]])

print("Prediction:", prediction[0])

if prediction[0] == 1:
    print("Customer Will Buy Product")
else:
    print("Customer Will Not Buy Product")
```

## 8. Evaluating Model Performance

Accuracy measures how many predictions are correct.

```python
train_predictions = model.predict(X)

accuracy = accuracy_score(y, train_predictions)

print("Accuracy:", accuracy * 100, "%")
```

## 9. Understanding the Kernel

A Kernel helps SVM handle complex datasets.

Common Kernels:

- Linear
- Polynomial
- RBF (Radial Basis Function)
- Sigmoid

In this notebook, we use the Linear Kernel.

## 10. Saving the Model

The trained model can be saved and reused later.

```python
joblib.dump(model, "svm_model.pkl")

print("Model Saved Successfully")
```

## 11. Loading the Saved Model

The saved model can be loaded without retraining.

```python
loaded_model = joblib.load("svm_model.pkl")

result = loaded_model.predict([[32, 58000]])

print(result)
```
