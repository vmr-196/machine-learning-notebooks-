# Decision Tree Classifier

A beginner-friendly implementation of Decision Tree using Scikit-Learn.

**Objective:** Predict whether a customer will buy a car based on age and income.

## 1. Introduction

Decision Tree is a supervised machine learning algorithm used for classification and regression tasks.

It works like a flowchart where decisions are made by asking a series of questions.

### Applications

- Loan Approval
- Disease Prediction
- Customer Churn Prediction
- Fraud Detection
- Product Recommendations

## 2. How Decision Trees Work

A Decision Tree splits data into smaller groups based on conditions.

Example:

Income > ₹50,000?

Yes → Buy Car

No → Don't Buy Car

The algorithm automatically finds the best questions to ask.

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree

from sklearn.metrics import accuracy_score
import joblib
```

## 3. Creating the Dataset

Dataset Features:

- Age
- Income

Target:

- Buy Car

0 = No
1 = Yes

```python
X = np.array([
    [22, 25000],
    [25, 30000],
    [28, 35000],
    [35, 60000],
    [40, 65000],
    [45, 70000],
    [23, 27000],
    [38, 62000]
])

y = np.array([
    0,
    0,
    0,
    1,
    1,
    1,
    0,
    1
])

print("Dataset Loaded Successfully")
```

## 4. Creating the Decision Tree Model

Now we create the Decision Tree classifier.

```python
model = DecisionTreeClassifier(random_state=42)
```

## 5. Training the Model

The fit() method trains the model using the dataset.

```python
model.fit(X, y)

print("Model Trained Successfully")
```

## 6. Making Predictions

Let's predict whether a customer aged 30 with an income of ₹55,000 will buy a car.

```python
prediction = model.predict([[30, 55000]])

print("Prediction:", prediction[0])

if prediction[0] == 1:
    print("Customer Will Buy Car")
else:
    print("Customer Will Not Buy Car")
```

## 7. Evaluating Model Performance

Accuracy measures how many predictions are correct.

```python
train_predictions = model.predict(X)

accuracy = accuracy_score(y, train_predictions)

print("Accuracy:", accuracy * 100, "%")
```

## 8. Visualizing the Decision Tree

The tree structure shows how the model makes decisions.

```python
plt.figure(figsize=(10,6))

plot_tree(
    model,
    feature_names=["Age", "Income"],
    class_names=["No", "Yes"],
    filled=True
)

plt.show()
```

## 9. Saving the Model

The trained model can be saved for future use.

```python
joblib.dump(model, "decision_tree.pkl")

print("Model Saved Successfully")
```

## 10. Loading the Saved Model

The saved model can be loaded and used for predictions.

```python
loaded_model = joblib.load("decision_tree.pkl")

result = loaded_model.predict([[32, 58000]])

print(result)
```
