# Apriori Algorithm

A beginner-friendly implementation of the Apriori Algorithm using Python.

**Objective:** Discover which gaming accessories are frequently purchased together.

## 1. Introduction

Apriori is an Unsupervised Machine Learning algorithm used for Association Rule Mining.

It helps discover relationships between products that are frequently purchased together.

### Applications

- Product Recommendations
- Market Basket Analysis
- E-commerce Analytics
- Inventory Management
- Customer Behavior Analysis

Example:

Gaming Mouse → Gaming Keyboard

This means customers purchasing a Gaming Mouse often purchase a Gaming Keyboard as well.

```python
import pandas as pd

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

## 3. Creating the Dataset

Each list below represents one customer's purchase transaction from a gaming store.

```python
transactions = [
    ['Gaming Mouse', 'Gaming Keyboard', 'Mouse Pad'],
    ['Gaming Headset', 'Gaming Mouse'],
    ['Gaming Keyboard', 'Mouse Pad'],
    ['Gaming Mouse', 'Gaming Headset', 'Mouse Pad'],
    ['Gaming Chair', 'Gaming Desk'],
    ['Gaming Mouse', 'Gaming Keyboard'],
    ['Gaming Headset', 'Gaming Keyboard'],
    ['Gaming Mouse', 'Gaming Chair'],
    ['Gaming Mouse', 'Gaming Keyboard', 'Gaming Headset'],
    ['Gaming Desk', 'Gaming Chair']
]

transactions
```

## 4. Understanding Support

Support measures how often an itemset appears in the dataset.

Formula:

Support(A → B) =
(Number of Transactions containing A and B)
/
(Total Transactions)

Higher support means the itemset occurs frequently.

### Example

Suppose:

Total Transactions = 10

Gaming Mouse and Gaming Keyboard together = 3

Support = 3 / 10 = 0.30

Meaning:

30% of all customers purchased both products together.

## 5. Understanding Confidence

Confidence measures how often the rule is correct.

Formula:

Confidence(A → B) =
(Number of Transactions containing A and B)
/
(Number of Transactions containing A)

Higher confidence means stronger predictive power.

### Example

Gaming Mouse buyers = 6

Gaming Mouse and Gaming Keyboard together = 3

Confidence = 3 / 6 = 0.50

Meaning:

50% of customers who bought a Gaming Mouse also bought a Gaming Keyboard.

### Lift Interpretation

| Lift Value | Meaning |
|------------|----------|
| Lift < 1 | Negative Relationship |
| Lift = 1 | No Relationship |
| Lift > 1 | Positive Relationship |
| Lift > 2 | Strong Relationship |
| Lift > 3 | Very Strong Relationship |

### Example

Confidence(Mouse → Keyboard) = 0.50

Support(Keyboard) = 0.40

Lift = 0.50 / 0.40

Lift = 1.25

Meaning:

Customers purchasing a Gaming Mouse are 1.25 times more likely to purchase a Gaming Keyboard than a random customer.

## 7. Converting Transactions into Tabular Format

Apriori requires data in one-hot encoded format.

```python
te = TransactionEncoder()

te_array = te.fit(transactions).transform(transactions)

df = pd.DataFrame(
    te_array,
    columns=te.columns_
)

df
```

## 8. Finding Frequent Itemsets

We will use a minimum support value of 30%.

Only itemsets appearing in at least 30% of transactions will be selected.

```python
frequent_itemsets = apriori(
    df,
    min_support=0.3,
    use_colnames=True
)

frequent_itemsets
```

## 9. Generating Association Rules

Association rules help identify relationships between products.

```python
rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.5
)

rules
```

## 10. Displaying Important Metrics

Let's view the most useful information from the generated rules.

```python
rules[
    [
        'antecedents',
        'consequents',
        'support',
        'confidence',
        'lift'
    ]
].sort_values(
    by='lift',
    ascending=False
)
```

## 11. Rule Analysis

The strongest rules generally have:

- High Support
- High Confidence
- High Lift

These rules are useful for recommendation systems and sales optimization.

## Example Interpretation

Rule:

Gaming Mouse → Gaming Keyboard

Interpretation:

- Support tells us how common the combination is.
- Confidence tells us how often the rule is correct.
- Lift tells us how strong the relationship is compared to random chance.

```python
rules.to_csv(
    "association_rules.csv",
    index=False
)

print("Rules Saved Successfully")
```
